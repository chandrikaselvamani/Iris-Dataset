#### IMPORTING IMPORTANT LIBRARIES

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib qt
from warnings import filterwarnings
```

#### LOADING THE DATASET

```python
iris=load_iris()
iris
```

```python
irisdf=pd.DataFrame(data=iris.data,columns=iris.feature_names)
irisdf
```

```python
irisdf['species']=iris.target
irisdf.head()
```

```python
def recode(x):
    if x==0:
        return 'Setosa'
    elif x==1:
        return 'Versicolor'
    else:
        return 'Virginica'

```

```python
irisdf['species']=irisdf['species'].apply(recode)
irisdf
```

#### CHECKING FOR MISSING VALUES

```python
irisdf.isnull().sum()
```

#### SPLITTING OF INDEPENDENT VARIABLES AND RESPONSE VARIABLES

```python
x=irisdf.iloc[:,:4]
x
```

```python
y=irisdf.iloc[:,-1]
y
```

#### SPLITTING OF TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(random_state=1)
```

#### FITTING THE MODEL

```python
classifier.fit(train_x,train_y)
```

```python
pred_train=classifier.predict(train_x)
pred_test=classifier.predict(test_x)
```

#### CHECKING FOR THE ACCURACY OF THE MODEL

```python
from sklearn.metrics import classification_report
```

```python
print (classification_report(train_y,pred_train))
```

```python
print (classification_report(test_y,pred_test))
```
